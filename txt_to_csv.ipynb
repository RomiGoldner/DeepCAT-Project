```python
def create_csv(txt_files, path, joined_df, all_samples_df):
    # parse through txt files
    for text_file_name in txt_files:
      sample_name = text_file_name

      # convert text_file to csv
      txt_file_path = os.path.join(path, text_file_name)
      # set delimeter type
      delimeter_type = "\t"
      # read txt file as df
      df = pd.read_csv(txt_file_path, delimiter=delimeter_type)

      # drop unecessary columns
      df.drop(['vMaxResolved', 'frequencyCount....', 'Group'], axis=1, inplace=True)

      # edit sample name before adding to csv
      sample_name_edit = sample_name.split('-')[1]
      sample_name_edit = sample_name_edit.split('.')[0]
      # add 'sample' column to df (name_only)
      df['sample'] = sample_name_edit

      # merge df_subset with joined_df
      # import pdb; pdb.set_trace()
      frames = [joined_df, df]
      result_df = pd.concat(frames)
      joined_df = result_df
      
    both_df = [all_samples_df, result_df]
    all_samples_df = pd.concat(both_df)
    return all_samples_df
```

```python
# parse through mixcr_output folder all the way to the txt files
root_path = '/content/drive/MyDrive/Associated_Datasets'

# general csv file name for combined csv
csv_file_name = "sequences_samples_csv_file.csv"

# create joined_df dataframe
joined_df = pd.DataFrame(columns=["aminoAcid", "sample"])
all_samples_df = pd.DataFrame(columns=["aminoAcid", "sample"])


for root, dirs, files in os.walk(root_path):
  txt_files = [f for f in files if f.endswith('.txt')]

  if len(txt_files) == 0:
    continue

  path = Path(root)
  if path.parent.name == "Other_Healthy":
    print(path.parent.name)
    continue
  
  if path.parent.name == "Pancreatic":
    print(path.parent.name)
    continue

  if path.parent.name == "Early_stage_BRCA":
    print(path.parent.name)
    csv_file_name = "brca_samples.csv"

  if path.parent.name == "Healthy_donors":
    print(path.parent.name)
    csv_file_name = "healthy_samples.csv"
  
  # parse through txt files
  for text_file_name in txt_files:
    sample_name = text_file_name

    # convert text_file to csv
    txt_file_path = os.path.join(path, text_file_name)
    # set delimeter type
    delimeter_type = "\t"
    # read txt file as df
    df = pd.read_csv(txt_file_path, delimiter=delimeter_type)

    # drop unecessary columns
    df.drop(['vMaxResolved', 'frequencyCount....', 'Group'], axis=1, inplace=True)

    # edit sample name before adding to csv
    sample_name_edit = sample_name.split('-')[1]
    sample_name_edit = sample_name_edit.split('.')[0]
    # add 'sample' column to df (name_only)
    df['sample'] = sample_name_edit

    # merge df_subset with joined_df
    # import pdb; pdb.set_trace()
    frames = [joined_df, df]
    result_df = pd.concat(frames)
    joined_df = result_df
    
  both_df = [all_samples_df, result_df]
  all_samples_df = pd.concat(both_df)



  # export joined_df (all sequences) to csv
  csv_path = '/content/drive/MyDrive/Associated_Datasets/' + csv_file_name
  print(csv_path)
  df_to_csv = create_csv(txt_files, path, joined_df, all_samples_df)
  print(len(df_to_csv))
  df_to_csv.to_csv(csv_path)
```
